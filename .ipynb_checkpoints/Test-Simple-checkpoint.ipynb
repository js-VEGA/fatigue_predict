```python
import tensorflow as tf
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior() 
import numpy as np
```

```python
X = tf.placeholder(tf.float32,shape=[None,1])
Y = tf.placeholder(tf.float32,shape=[None,1])
W = tf.Variable(tf.random_normal([1,1],mean = -3),name="weight")
b = tf.Variable(tf.random_normal([1]),name="bias")
```

```python
hypothesis = tf.matmul(X,W)+b
```

```python
saver = tf.train.Saver()
model = tf.global_variables_initializer()
```

```python
sleep_time = float(input('수면 시간: '))
```

```python
with tf.Session() as sess:

    sess.run(model)

    save_path = "./simple/saved.cpkt"

    saver.restore(sess, save_path)



    # 사용자의 입력 값을 이용해 배열을 만듭니다.

    data = ((sleep_time,),)

    arr = np.array(data, dtype=np.float32)



    # 예측을 수행한 뒤에 그 결과를 출력합니다.

    x_data = arr[0:1]

    dict = sess.run(hypothesis, feed_dict={X: x_data})

    print(dict[0])

    #먼저 원래 파일을 훈련 시킨다음에 이 단락만 실행시켜야 결과가 나옴
    #안그러면 오류가 난다.
    #한번에 한번씩,여러번 하면 오류 생김->다시 탭 닫고 실행해야함.
```
