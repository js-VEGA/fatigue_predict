```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
f = open ('correct_data_num.csv')
data = csv.reader(f)
```

```python
#x는 잠잔 시간,y는 휴대폰 사용 시간,z는 피로도
x = []
y = []
z = []

for row in data:
    int_row = [int(i) for i in row]
    print(int_row)
    x.append(int_row[0])
    y.append(int_row[1])
    z.append(int_row[2])
    
print(x)
print(y)
print(z)
    
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z,color = "r")

f.close
```
